<div style="font-family:Georgia;line-height:25px">
<strong>Data Normalization</strong>

This is an important pre-processing step for neural networks:
</div>


```python
# normalize data
df2 = df.copy(deep=True)
scaler = MinMaxScaler(feature_range=(0,15)).fit(df2.Low.values.reshape(-1,1))
df2['Open'] = scaler.transform(df2.Open.values.reshape(-1,1))
df2['High'] = scaler.transform(df2.High.values.reshape(-1,1))
df2['Low'] = scaler.transform(df2.Low.values.reshape(-1,1))
df2['Close'] = scaler.transform(df2.Close.values.reshape(-1,1))
data = df2[['Open','High','Low', 'Close']].values
data.shape
```

```python
data
```

<div style="font-family:Georgia;line-height:25px">
<strong>Preparing Data for Time Series</strong>

In order to prepare the dataset for stock price prediction, we need to frame it as a time series problem. We will use the price from the previous 10 days to predict the price on the next day. Let‚Äôs 
create the sequences accordingly:
</div>

```python
seq_len=11
sequences=[]
for index in range(len(data) - seq_len + 1): 
    sequences.append(data[index: index + seq_len])
sequences= np.array(sequences)
```

<div style="font-family:Georgia;line-height:25px">
<strong>Splitting the Data</strong>

We divide the entire dataset into three parts. 80% for the training set, 10% for the validation set and the remaining 10% for the test set:
</div>

```python
valid_set_size_percentage = 10 
test_set_size_percentage = 10 

valid_set_size = int(np.round(valid_set_size_percentage/100*sequences.shape[0]))  
test_set_size  = int(np.round(test_set_size_percentage/100*sequences.shape[0]))
train_set_size = sequences.shape[0] - (valid_set_size + test_set_size)

x_train = sequences[:train_set_size,:-1,:]
y_train = sequences[:train_set_size,-1,:]
    
x_valid = sequences[train_set_size:train_set_size+valid_set_size,:-1,:]
y_valid = sequences[train_set_size:train_set_size+valid_set_size,-1,:]
    
x_test = sequences[train_set_size+valid_set_size:,:-1,:]
y_test = sequences[train_set_size+valid_set_size:,-1,:]
```

# <p style="background-image: url(https://i.postimg.cc/CLNcqSzH/finance-and-business-background-bar-chart-and-candlestick-chart-show-stock-market-trading-price-vect.jpg);font-family:newtimeroman;font-size:100%;color:white;text-align:center;border-radius:15px 50px; padding:7px">5.1. Model 1 (Vanilla LSTM)</p>

<a class="btn" href="#home">üè† Tabel of Contents</a>

<div style="font-family:Georgia;line-height:25px">
A Vanilla LSTM is an LSTM model that has a single hidden layer of LSTM units, and an output layer used to make a prediction</div>

```python
# build model 1 (Vanilla LSTM)
model1 = Sequential()
model1.add(LSTM(units=50, input_shape=(seq_len-1,4), kernel_regularizer=regularizers.l2(0.001)))
model1.add(Dense(units=4))

model1.summary()
```

```python
# compile model 1
model1.compile(optimizer='adam', loss='mean_squared_error', metrics=['msle'])
```

```python
model1_hist  = model1.fit(
    x_train,
    y_train,
    epochs=50,
    batch_size=64,
    validation_data=(x_valid, y_valid),
)
```

```python
def plot_model_history(model_hist):
    """
    Plot diagram of accuracy and loss for train and validation set
        model_hist: results of model
    """
    
    fig, axs = plt.subplots(2, 1, figsize=(10, 12))
    
    # Plot train and validation accuracy
    axs[0].plot(model_hist.history['msle'], color="red", marker="o")
    axs[0].plot(model_hist.history['val_msle'], color="blue", marker="h")
    axs[0].set_title('MSLE Comparison between Train & Validation Set')
    axs[0].set_ylabel('msle')
    axs[0].set_xlabel('Epoch')
    axs[0].legend(['Train', 'Validation'], loc="lower right")
    
    # Plot train and validation loss
    axs[1].plot(model_hist.history['loss'], color="red", marker="o")
    axs[1].plot(model_hist.history['val_loss'], color="blue", marker="h")
    axs[1].set_title('Loss Comparison between Train & Validation Set')
    axs[1].set_ylabel('Loss')
    axs[1].set_xlabel('Epoch')
    axs[1].legend(['Train', 'Validation'], loc="upper right")

    plt.tight_layout()
    plt.show()
```

```python
plot_model_history(model1_hist)
```

```python
predictions1 = model1.predict(x_test)
# Calculate RMSE
rmse = np.sqrt(mean_squared_error(y_test, predictions1))
print(f'RMSE of Model 1: {rmse:0.3}')

predictions1 = scaler.inverse_transform(predictions1)
```

```python
model1.evaluate(x_test, y_test)
```

```python
# Plotting the data
def plot_predictions(predictions, idx, data_source=df, test_range=len(df) - len(y_test)):

    plt.figure(figsize=(15, 8))
    plt.subplot(2, 1, 1,)    
    plt.title('Model', fontsize = 18)
    plt.xlabel('Date', fontsize = 14)
    plt.ylabel('Closing Price', fontsize = 14)
    plt.plot(df.iloc[:test_range, idx])
    plt.plot(df.iloc[test_range:, idx])
    plt.plot(df[test_range:].index, predictions[:, idx])

    # dates = pd.date_range('2023-01-04', periods=133, freq='D')

    # Define the window start and end dates
    window_start = df[test_range:].index[0]
    window_end = df[test_range:].index[-1]

    # Define width and height
    width = window_end - window_start
    height = max(predictions[:, idx]) - min(predictions[:, idx]) + 30

    # Create a Rectangle object
    rectangle = Rectangle((window_start, min(predictions[:, idx])-10), width, height, fill=False, edgecolor='red', linestyle='dashed')

    # Add the Rectangle to the plot
    plt.gca().add_patch(rectangle)
    
    plt.figure(figsize=(8, 8))
    plt.subplot(2, 1, 2)
    plt.title('Model', fontsize = 14)
    plt.xlabel('Date', fontsize = 12)
    plt.ylabel('Closing Price', fontsize = 12)
    plt.xticks(rotation=45)
    plt.plot(df.iloc[test_range:, idx], color='orange', label='Close')
    plt.plot(df.iloc[test_range:].index, predictions[:, idx], color='green', label='Pred')
    plt.legend(['test', 'Prediction'], loc = 'lower right')

    plt.tight_layout()
    plt.show()
```

```python
plot_predictions(predictions1, 2, data_source=df)
```

<div style="font-family:Georgia;line-height:25px">
Multiple hidden LSTM layers can be stacked one on top of another in what is referred to as a Stacked LSTM model.</div>

```python
# build model2
model2 = Sequential()
model2.add(LSTM(units=50, return_sequences=True, input_shape=(seq_len-1, 4)))
model2.add(LSTM(units=50))
model2.add(Dropout(0.2))
model2.add(Dense(units=4))

model2.summary()
```

Model 3 (Bidirectional LSTM)</p>

<a class="btn" href="#home">üè† Tabel of Contents</a>

<div style="font-family:Georgia;line-height:25px">

On some sequence prediction problems, it can be beneficial to allow the LSTM model to learn the input sequence both forward and backwards and concatenate both interpretations.

This is called a Bidirectional LSTM.

We can implement a Bidirectional LSTM for univariate time series forecasting by wrapping the first hidden layer in a wrapper layer called Bidirectional.
</div>


```python
# build model3
model3 = Sequential()
model3.add(Bidirectional(LSTM(units=50, input_shape=(seq_len-1, 4), kernel_regularizer=regularizers.l2(0.001))))
model3.add(Dense(units=4))
```

```python
# compile model3
model3.compile(optimizer='adam', loss='mse', metrics=['msle'])
```

```python
model3_hist  = model3.fit(
    x_train,
    y_train,
    epochs=50,
    batch_size=64,
    validation_data=(x_valid, y_valid),
)
```

```python
plot_model_history(model3_hist)
```
